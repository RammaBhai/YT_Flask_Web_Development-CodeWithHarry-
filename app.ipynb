```python
from flask import Flask, request  # removed render_template since unused

app = Flask(__name__)


@app.route("/")
def home():
    return "<h1>Welcome to My Mini Flask App!</h1>"


@app.route("/greet", methods=["GET", "POST"])
def greet():
    if request.method == "POST":
        name = request.form.get("name", "Guest")
        return f"<h3>Hello, {name}! ðŸ‘‹</h3>"
    return """
        <form method="POST">
            Enter your name: <input type="text" name="name">
            <input type="submit" value="Greet Me">
        </form>
    """


if __name__ == "__main__":
    app.run(debug=True)
```
